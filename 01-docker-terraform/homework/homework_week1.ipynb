**Question 3. Counting short trips**

```python
# Filter trips in November 2025
november_trips = df[
    (df["lpep_pickup_datetime"] >= "2025-11-01") &
    (df["lpep_pickup_datetime"] < "2025-12-01")
]

november_trips.head()
```

```python
# Count trips with distance <= 1 mile
count_short_trips = november_trips[november_trips["trip_distance"] <= 1].shape[0]

count_short_trips
```

**Question 4. Longest trip for each day**

```python
# Filter valid trips (exclude outliers)
valid_trips = df[df["trip_distance"] < 100].copy()

# Create pickup date column (date only, no time)
valid_trips["pickup_date"] = valid_trips["lpep_pickup_datetime"].dt.date

# Find max trip distance per pickup day
daily_max_distance = (
    valid_trips
    .groupby("pickup_date")["trip_distance"]
    .max()
)
```

```python
# Find the pickup day with the longest trip distance
longest_trip_day = daily_max_distance.idxmax()
longest_trip_distance = daily_max_distance.max()

longest_trip_day, longest_trip_distance
```

**Question 5. Biggest pickup zone**

```python
taxi_zone = pd.read_csv("taxi_zone_lookup.csv")
taxi_zone.head()
```

```python
taxi_zone.info()
```

```python
# Filter trips on Nov 18, 2025 and calculate total_amount per pickup location
nov18 = df[
    (df["lpep_pickup_datetime"] >= "2025-11-18") &
    (df["lpep_pickup_datetime"] < "2025-11-19")
].copy()

pickup_sum = (
    nov18.groupby("PULocationID")["total_amount"]
    .sum()
    .reset_index()
)
pickup_sum.head(10)
```

```python
# Join with taxi zone lookup table and select the zone with the highest total_amount
top_zone = (
    pickup_sum
    .merge(taxi_zone, left_on="PULocationID", right_on="LocationID", how="left")
    .sort_values("total_amount", ascending=False)
    .head(1)
)

top_zone[["Zone", "Borough", "total_amount"]]
#top_zone
```

**Quesion 6. Largest tip**

```python
# 1. Get LocationID for "East Harlem North"
east_harlem_id = taxi_zone.loc[taxi_zone['Zone'] == 'East Harlem North', 'LocationID'].values[0]

# 2. Filter trips from East Harlem North in Nov 2025
mask = (
    (df['PULocationID'] == east_harlem_id) &
    (df['lpep_pickup_datetime'].dt.year == 2025) &
    (df['lpep_pickup_datetime'].dt.month == 11)
)
df_filtered = df[mask]
```

```python
df_filtered.head(5)
```

```python
# 3. Find the row with the largest tip
max_tip_row = df_filtered.loc[df_filtered['tip_amount'].idxmax()]

# 4. Get the drop-off zone name
dropoff_zone_id = max_tip_row['DOLocationID']
dropoff_zone_name = taxi_zone.loc[taxi_zone['LocationID'] == dropoff_zone_id, 'Zone'].values[0]

print("Drop-off zone with the largest tip:", dropoff_zone_name)
```
